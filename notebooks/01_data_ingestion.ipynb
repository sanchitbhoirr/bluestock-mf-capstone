```python
import pandas as pd
import os

raw_data_path = "../data/raw/"

csv_files = [
    "01_fund_master.csv",
    "02_nav_history.csv",
    "03_aum_by_fund_house.csv",
    "04_monthly_sip_inflows.csv",
    "05_category_inflows.csv",
    "06_industry_folio_count.csv",
    "07_scheme_performance.csv",
    "08_investor_transactions.csv",
    "09_portfolio_holdings.csv",
    "10_benchmark_indices.csv"    
]

print("Starting Data Integestion.. \n" + "="*50)
```

```python
for file in csv_files:
    file_path = os.path.join(raw_data_path, file)
    
    if os.path.exists(file_path):
        print(f"--- Loading {file} ---")
        df = pd.read_csv(file_path)
        print(f"Shape: {df.shape}")
        print("Data types: ")
        print(df.dtypes)
        print("\nFirst 3 Rows: ")
        print(df.head(3))
        print("="*60, "\n")
    else:
        print(f"Files not found: {file_path}")
```

```python
import pandas as pd

fund_master = pd.read_csv('../data/raw/01_fund_master.csv')
nav_history = pd.read_csv('../data/raw/02_nav_history.csv')

fund_master.columns = fund_master.columns.str.strip()
nav_history.columns = nav_history.columns.str.strip()

print("=== DATA QUALITY SUMMARY: AMFI CODE VALIDATION ===\n")

master_codes = set(fund_master['amfi_code'].unique())

nav_col = 'scheme_code' if 'scheme_code' in nav_history.columns else 'amfi_code'

if nav_col in nav_history.columns:
    nav_codes = set(nav_history[nav_col].unique())
    
    missing_codes = master_codes - nav_codes
    
    print(f"Total Funds in Fund Master: {len(master_codes)}")
    print(f"Total Funds with Data in NAV History: {len(nav_codes)}\n")

    if not missing_codes:
        print("SUCCESS: All AMFI codes in fund_master perfectly match and exist in nav_history.")
        print("Summary: Data quality is excellent. No missing historical records for the selected funds.")
    else:
        print(f"WARNING: {len(missing_codes)} AMFI codes are missing in nav_history!")
        print(f"Missing Codes: {list(missing_codes)}")
        print("Summary: Data quality issue detected. Some master funds do not have historical NAV data.")
else:
    print(f"Error: Cannot find scheme code column in nav_history. Columns are: {nav_history.columns.tolist()}")
```
